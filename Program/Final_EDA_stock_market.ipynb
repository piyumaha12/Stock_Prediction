```python
# Converting to datetime datatype
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['date_year'] = df['Date'].dt.year

df.sample(5)
```

```python
df['date_month_no'] = df['Date'].dt.month

df.head()
```

```python
df['date_month_name'] = df['Date'].dt.month_name()

df.head()
```

```python
df['date_day'] = df['Date'].dt.day

df.head()
```

```python
# day of week
df['date_dow'] = df['Date'].dt.dayofweek

df.head()
```

```python
# day of week - name

df['date_dow_name'] = df['Date'].dt.day_name()
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.boxplot(x= 'Open', data=df1)
```

```python
plt.figure(figsize = (10,8))
sns.catplot(x = 'Date' , y = 'Open', data = df1)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x=df1['Date'],y=df1['Open'],color='blue')
plt.xticks(rotation='vertical')
plt.xlabel('date')
plt.ylabel('Open stock price')
plt.title("Day wise stock price",fontsize=16)

plt.tight_layout(pad=0.5)
plt.show()
```

```python
sns.distplot(df1['Open'])
```

```python
sns.kdeplot(df1['Open'],shade=True, color='b')
```

# Log Transformation

```python
import warnings
warnings.filterwarnings('ignore')
log = np.log(df1['Open'])
sns.distplot(log)
fig = plt.figure()
```

```python
# Square root transformation

sqrt = np.sqrt(df1['Open'])
sns.distplot(sqrt)
fig = plt.figure()
```

```python
# Reciprocal Transformation

rec = 1/(df1['Open']+1)
sns.distplot(rec)
fig = plt.figure()
```

```python
# Power Transformation:Box-Cox transformation

from scipy import stats
pt = stats.boxcox(df1['Open'])[0]
sns.distplot(pt)
fig = plt.figure()
```
